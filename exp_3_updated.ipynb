```
# Apply log transformation method 
c = 255 / np.log(1 + np.max(ima)) 
log_image = c * (np.log(ima + 1)) 
   
# Specify the data type so that 
# float value will be converted to int 
log_image = np.array(log_image, dtype = np.uint8) 
   
# Display both images 
cv2_imshow(log_image)
```

```
final_frame = cv.hconcat((ima, log_image))
cv2_imshow(final_frame)
```

```
ima_gray = cv.imread('exp 2.2.jpg', cv.IMREAD_GRAYSCALE)
from google.colab.patches import cv2_imshow
cv2_imshow(ima_gray)
```

```
# Apply log transformation method 
c = 255/5
log_image = c * (np.log(ima_gray + 1)) 
   
# Specify the data type so that 
# float value will be converted to int 
log_image = np.array(log_image, dtype = np.uint8) 
   


cv2_imshow(log_image)
```

```
final_frame = cv.hconcat((ima_gray, log_image))
cv2_imshow(final_frame)
```

```
for gamma in [0.1, 0.5, 1.2, 2.2]: 
      
    # Apply gamma correction. 
    gamma_corrected = np.array(255*(ima / 255) ** gamma, dtype = 'uint8') 
  
    cv2_imshow(gamma_corrected) 
 
```

```

img = cv.imread('exp 2.2.jpg')
imgs = [255*((img& (1<<i)) >>i) for i in range(8)]

#Concatenate these images for ease of display using cv2.hconcat()
finalr = cv.hconcat([imgs[0],imgs[1],imgs[2],imgs[3]])
finalv =cv.hconcat([imgs[4],imgs[5],imgs[6],imgs[7]])

 
# Vertically concatenate
final = cv.vconcat([finalr,finalv])
 
# Display the images
cv2_imshow(final)
```
