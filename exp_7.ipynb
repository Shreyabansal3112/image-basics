```
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
```

```
ima = cv.imread('jpg.jpg')
from google.colab.patches import cv2_imshow
cv2_imshow(ima)
```

```
img = cv.cvtColor(ima, cv.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```
#CV_8U - 8-bit unsigned integers ( 0..255 )

#CV_8S - 8-bit signed integers ( -128..127 )

#CV_16U - 16-bit unsigned integers ( 0..65535 )

#CV_16S - 16-bit signed integers ( -32768..32767 )

#CV_32S - 32-bit signed integers ( -2147483648..2147483647 )

#CV_32F - 32-bit floating-point numbers ( -FLT_MAX..FLT_MAX, INF, NAN )

#CV_64F - 64-bit floating-point numbers ( -DBL_MAX..DBL_MAX, INF, NAN )


```

```
laplacian = cv.Laplacian(img,cv.CV_16S)
cv2_imshow(laplacian)
```

```
laplacian = cv.Laplacian(img,-1)
cv2_imshow(laplacian)
```

```
kernelx = np.array([[-1, 0], [0, 1]], dtype=int)
kernely = np.array([[0, -1], [1, 0]], dtype=int)

x = cv.filter2D(img, cv.CV_16S, kernelx)
y = cv.filter2D(img, cv.CV_16S, kernely)
absX = cv.convertScaleAbs(x)
absY = cv.convertScaleAbs(y)
Roberts = cv.addWeighted(absX, 0.5, absY, 0.5, 0)
cv2_imshow(Roberts)
```

```
kernelx = np.array([[-1,-1,-1],[0,0,0],[1,1,1]],dtype=int)
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]],dtype=int)
x = cv.filter2D(img, cv.CV_16S, kernelx)
y = cv.filter2D(img, cv.CV_16S, kernely)

# Turn uint8, image fusion
absX = cv.convertScaleAbs(x)
absY = cv.convertScaleAbs(y)
Prewitt = cv.addWeighted(absX, 0.5, absY, 0.5, 0)
cv2_imshow(Prewitt)
```

```
sobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=5)  # x
sobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)  # y
cv2_imshow(sobelx)
```

```
cv2_imshow(sobely)
```

```
Sobel = cv.addWeighted(sobelx, 0.5, sobely, 0.5, 0)
cv2_imshow(Sobel)
```

```
sobelall = cv.Sobel(img,cv.CV_64F,1,1,ksize=5)
cv2_imshow(sobelall)
```

```
edges = cv.Canny(img,100,200)
```

```
cv2_imshow(edges)
```

```
edges = cv.Canny(img,10,300)
cv2_imshow(edges)
```
