```
import cv2 as cv 
import matplotlib.pyplot as plt
```

```
from google.colab.patches import cv2_imshow
ima = cv.imread('test-exp-2.jpg', cv.IMREAD_UNCHANGED)
cv2_imshow(ima)
```

```
ima = cv.imread('test-exp-2.jpg', cv.IMREAD_UNCHANGED)
imaRGB=cv.cvtColor(ima,cv.COLOR_BGR2RGB)
cv2_imshow(imaRGB)
```

```
ima = cv.imread('test-exp-2.jpg', cv.IMREAD_UNCHANGED)
imaHSV=cv.cvtColor(ima,cv.COLOR_BGR2HSV)
cv2_imshow(imaHSV)
```

```
src = cv.imread('test-exp-2.jpg',-1)
R = src[:, :, 2]
G = src[:, :, 1]
B = src[:, :, 0]
```

```
cv2_imshow(R)
```

```
cv2_imshow(B)
```

```
cv2_imshow(G)
```

```
plt.imshow(R,cmap=plt.cm.Reds)
```

```
plt.imshow(B,cmap=plt.cm.Blues)
```

```
import numpy as np
src = cv.imread('test-exp-2.jpg',-1)
src[:,:,2] = np.zeros([src.shape[0], src.shape[1]])
cv2_imshow(src)
```

```
src = cv.imread('test-exp-2.jpg',-1)
src[:,:,1] = np.zeros([src.shape[0], src.shape[1]])
cv2_imshow(src)
```

```
src = cv.imread('test-exp-2.jpg',-1)
src[:,:,0] = np.zeros([src.shape[0], src.shape[1]])
cv2_imshow(src)
```

```
src = cv.imread('gray scale.png',-1)
cv2_imshow(src)
```

```
im_gray  = cv.imread( "gray scale.png" , cv.IMREAD_GRAYSCALE)
im_color  = cv.applyColorMap(im_gray, cv.COLORMAP_RAINBOW)
cv2_imshow(im_color)
```
