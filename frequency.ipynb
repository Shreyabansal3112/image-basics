```python
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
ima = cv.imread('p2.jpg')
from google.colab.patches import cv2_imshow
cv2_imshow(ima)
ima.shape
```

```python
img = cv.imread('p2.jpg', cv.IMREAD_GRAYSCALE)
cv2_imshow(img)
img.shape
```

```python
f = np.fft.fft2(ima)
cv2_imshow(f)
```

```python
fshift = np.fft.fftshift(f)
cv2_imshow(fshift)
```

```python
rows, cols,band= ima.shape
crow,ccol = rows//2 , cols//2
fshift[crow-30:crow+31, ccol-30:ccol+31] = 0
cv2_imshow(fshift)
```

```python
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
cv2_imshow(np.int8(img_back))
```

```python
rows, cols,band= ima.shape
crow,ccol = rows//2 , cols//2
fshift[0:crow-30, 0:ccol-30] = 0
fshift[crow+31:rows, ccol+31:cols] = 0
cv2_imshow(fshift)
```

```python
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
cv2_imshow(np.int8(img_back))
```

```python
f = np.fft.fft2(img)
cv2_imshow(f)
```

```python
fshift = np.fft.fftshift(f)
cv2_imshow(fshift)
```

```python
rows, cols,band= ima.shape
crow,ccol = rows//2 , cols//2
fshift[crow-20:crow+21, ccol-20:ccol+21] = 0
cv2_imshow(fshift)
```

```python
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
cv2_imshow(np.int8(img_back))
```

```python
rows, cols = img.shape
crow,ccol = rows//2 , cols//2
# create a mask first, center square is 1, remaining all zeros
mask = np.zeros((rows,cols),np.uint8)
mask[crow-20:crow+20, ccol-20:ccol+20] = 1
mask.shape
```

```python

fshift = fshift*mask
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
plt.imshow(np.int8(img_back))
```
